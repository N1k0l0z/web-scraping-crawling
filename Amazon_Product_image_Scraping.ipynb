```python
def soap_func(page_source):
    soap = BeautifulSoup(page_source, 'html.parser')
    links = soap.find_all('a', attrs={'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    images = soap.find_all('img', attrs={'class': 's-image s-image-optimized-rendering'})
    link_list = []
    image_links = []
    for link in links:
        href = link.get('href')
        if href:
            link_list.append(f'https://www.amazon.com{href}')
    for image in images:
        src = image.get('src')
        if src:
            image_links.append(src)
    return link_list, image_links

def get_links(url):
    all_links = {"links": [], "images": []}
    chromedriver_autoinstaller.install()

    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument("user-agent=YourUserAgentHere")

    driver = webdriver.Chrome(options=options)
    driver.get(url)

    n = 1
    next_button = True
    while next_button:
        links, images = soap_func(driver.page_source)
        all_links["links"].extend(links)
        all_links["images"].extend(images)

        try:
            next_btn = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(@class, 's-pagination-next')]"))
            )
            actions = ActionChains(driver)
            actions.move_to_element(next_btn).perform()
            next_btn.click()
            sleep(5)
            WebDriverWait(driver, 10).until(EC.staleness_of(next_btn))
        except:
            next_button = False

    driver.quit()
    return all_links
```

```python
def download_images(image_urls):
    for idx, url in enumerate(image_urls):
        response = requests.get(url)
        filename = f"/content/drive/MyDrive/Final Project/Rolex/Images/image_{idx}.jpg"
        with open(filename, 'wb') as file:
            file.write(response.content)
download_images(data['images'].values)
```
